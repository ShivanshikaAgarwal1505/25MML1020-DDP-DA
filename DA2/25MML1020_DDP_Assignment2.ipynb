**Difference between Transformations and Actions in Spark**

In Apache Spark, operations on data are classified into transformations and actions, which are fundamental to its execution model.

**Transformations**

Transformations are operations that create a new DataFrame or RDD from an existing one. They are lazy, meaning they do not execute immediately. Instead, Spark builds a logical execution plan (DAG – Directed Acyclic Graph) and waits until an action is called.

Examples of Transformations

1. select()

Used to choose specific columns from a DataFrame.

Example: df.select("name", "age")

2. filter() / where()

Used to filter rows based on a condition.

Example: df.filter(df.age > 25)

3. groupBy()

Used to group data based on a column for aggregation.

Example: df.groupBy("department")

**Significance in Execution Planning**

* Transformations are lazily evaluated, so Spark can optimize the entire execution plan.
* Multiple transformations are combined into a single execution pipeline.
* Reduces unnecessary data movement and improves performance through optimization techniques like predicate pushdown and pipelining.


---


**Actions**

Actions are operations that trigger the execution of transformations. They return results to the driver program or write data to storage.

Examples of Actions
1. show()

Displays the DataFrame contents.

Example: df.show()

2. collect()

Returns all data to the driver as a list.

Example: df.collect()

3. count()

Returns the number of rows in the DataFrame.

Example: df.count()

**Significance in Execution Planning**
* Actions trigger the execution of the DAG created by transformations.
* Spark optimizes the plan before executing it.
* Without actions, transformations are not executed.
* Helps in efficient resource utilization by executing only required computations.

```python
# Import SparkSession
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("Spark Assignment 2") \
    .getOrCreate()


# Create sample dataset manually
data = [
    ("Alice", 25, "HR"),
    ("Bob", 30, "IT"),
    ("Charlie", 35, "IT"),
    ("David", 28, "Finance"),
    ("Eva", 40, "HR")
]

columns = ["Name", "Age", "Department"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Show original data (Action)
df.show()

# 1. Select specific columns (Transformation)
selected_df = df.select("Name", "Department")

# 2. Apply filter condition (Transformation)
filtered_df = df.filter(df.Age > 30)

# 3. Group by Department and count (Transformation)
grouped_df = df.groupBy("Department").count()

# Display results (Actions)
print("Selected Columns:")
selected_df.show()

print("Filtered Data (Age > 30):")
filtered_df.show()

print("Grouped Data (Count by Department):")
grouped_df.show()

# Stop Spark session
spark.stop()
```
