**Distributed Data Processing:**

Distributed data processing is a method of handling large datasets by dividing the data into smaller parts and processing them across multiple machines (nodes) in a cluster. Each node performs computations on its portion of the data in parallel, and the results are combined to produce the final output. This approach improves scalability, fault tolerance, and performance when dealing with big data.

**Apache Spark and Performance Improvement:**

Apache Spark is a distributed data processing framework designed to process large-scale data efficiently. It improves performance over traditional systems such as Hadoop MapReduce in the following ways:

1. In-Memory Computation:

Spark stores intermediate results in memory (RAM) rather than writing them to disk after every operation. This significantly reduces disk I/O and speeds up processing, especially for iterative tasks like machine learning and graph processing.

2. Lazy Evaluation:

Spark does not execute operations immediately. Instead, it builds a Directed Acyclic Graph (DAG) of transformations and optimizes the execution plan before running it. This reduces unnecessary computations.

3. Parallel Processing:

Spark distributes data across multiple nodes and executes tasks in parallel, leading to faster data processing.

4. Fault Tolerance:

Spark uses Resilient Distributed Datasets (RDDs), which can recover lost data using lineage information in case of node failures.

5. Efficient Execution Engine:

Compared to traditional systems like MapReduce (which involves multiple disk read/write cycles), Spark executes tasks in fewer stages, reducing overhead.

```python
# Import SparkSession
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("Spark Assignment 1") \
    .getOrCreate()

# Create data
data = [
    (1, "Alice", 25, 30000),
    (2, "Bob", 30, 40000),
    (3, "Charlie", 28, 25000),
    (4, "David", 35, 50000),
    (5, "Eve", 22, 20000)
]

# Define column names
columns = ["ID", "Name", "Age", "Salary"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Display schema
print("Schema:")
df.printSchema()

# Show all records
print("All Records:")
df.show()

# Filter employees with salary greater than 30000
filtered_df = df.filter(df["Salary"] > 30000)

# Show filtered records
print("Filtered Records (Salary > 30000):")
filtered_df.show()

# Stop Spark session
spark.stop()
```
